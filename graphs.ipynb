```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
femax = open('K3Femax.txt','rt')
femax = pd.DataFrame([[float(token) for token in line.split()] for line in femax.readlines()[::2]])
femin = open('K3Femin.txt','rt')
femin = pd.DataFrame([[float(token) for token in line.split()] for line in femin.readlines()[::2]])
plt.figure(figsize=(8,6))
plt.scatter(x=femax[1], y=femax[2], s=10, label='V_max')
plt.scatter(x=femin[1], y=femin[2], s=10, label='V_min')

plt.title('ВАХ Pt электрода')
plt.ylabel('$I,\ mА$')
plt.xlabel('$U, mV$')
plt.legend(loc='center right');

plt.savefig('saved_figure.png')
plt.show()
```

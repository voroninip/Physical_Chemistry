```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_excel(io='data.xlsx', sheet_name='Лист1')
df
```

```python
plt.figure(figsize=(10,5))

k, q = np.polyfit(list(df['x1'].iloc[0:3]), list(df['y1'].iloc[0:3]), deg=1)
sns.lineplot(data=df.iloc[0:3], x='x1', y='y1', label='$q_{max}:\ \ y = %.4f\\cdot x %.4f\ $' %(k, q))
sns.scatterplot(data=df.iloc[0:3], x='x1', y='y1')

k, q = np.polyfit(list(df['x2'].iloc[0:4]), list(df['y2'].iloc[0:4]), deg=1)
sns.lineplot(data=df.iloc[0:4], x='x2', y='y2', label='$q_1:\ \ y = %.4f\\cdot x + %.4f\ $' %(k, q))
sns.scatterplot(data=df.iloc[0:4], x='x2', y='y2')

k, q = np.polyfit(list(df['x3'].iloc[0:2]), list(df['y3'].iloc[0:2]), deg=1)
sns.lineplot(data=df.iloc[0:3], x='x3', y='y3', label='$q_2:\ \ y = %.4f\\cdot x %.4f\ $' %(k, q))
sns.scatterplot(data=df.iloc[0:3], x='x3', y='y3')

plt.xlabel('$N, W$')
plt.ylabel('$\\Delta T,K$')
plt.savefig('211_phys.png')
plt.show()
```

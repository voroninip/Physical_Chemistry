```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_excel(io='phys.xlsx', sheet_name='6M1000')
df = df[3:].drop(columns={'Column1', 'Column2', 'Unnamed: 2', 'Unnamed: 5', 'Unnamed: 6'})
df1 = df.rename(columns={'Unnamed: 3':'t', 'Unnamed: 4':'x', 'Unnamed: 7':'ln'})

df = pd.read_excel(io='phys.xlsx', sheet_name='2M300')
df = df[3:].drop(columns={'Столбец1', 'Столбец2', 'Unnamed: 2', 'Unnamed: 5'})
df2 = df.rename(columns={'Unnamed: 3':'t', 'Unnamed: 4':'x', 'Unnamed: 7':'ln'})

df = pd.read_excel(io='phys.xlsx', sheet_name='1200M3')
df = df[3:].drop(columns={'Column1', 'Column2', 'Unnamed: 2', 'Unnamed: 5'})
df3 = df.rename(columns={'Unnamed: 3':'t', 'Unnamed: 4':'x', 'Unnamed: 7':'ln'})

df = pd.read_excel(io='phys.xlsx', sheet_name='4M500')
df = df[3:].drop(columns={'Column1', 'Column2', 'Unnamed: 2', 'Unnamed: 5'})
df4 = df.rename(columns={'Unnamed: 3':'t', 'Unnamed: 4':'x', 'Unnamed: 7':'ln'})

df = pd.read_excel(io='phys.xlsx', sheet_name='5M750')
df = df[3:].drop(columns={'Column1', 'Column2', 'Unnamed: 2', 'Unnamed: 5'})
df5 = df.rename(columns={'Unnamed: 3':'t', 'Unnamed: 4':'x', 'Unnamed: 7':'ln'})

df1
```

```python
plt.figure(figsize=(10,6))

plt.scatter(x=df1['t'], y=df1['x'], s=2, label='5: 0.1M $\ NaOH$')
plt.scatter(x=df2['t'], y=df2['x'], s=0.3, label='1: 0.003M $\ NaOH$')
plt.scatter(x=df3['t'], y=df3['x'], s=0.7, label='2: 0.012M $\ NaOH$')
plt.scatter(x=df4['t'], y=df4['x'], s=1, label='3: 0.05M $\ NaOH$')
plt.scatter(x=df5['t'], y=df5['x'], s=2, label='4: 0.075M $\ NaOH$')

plt.xlabel('t, c')
plt.ylabel('$\\frac{D}{D_0}$')
plt.legend()
plt.savefig('1noln.png')
plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.scatter(x=df1['t'], y=df1['ln'], s=1, label='5: 0.1M $\ NaOH$')
plt.scatter(x=df2['t'], y=df2['ln'], s=0.3, label='1: 0.003 $\ NaOH$')
plt.scatter(x=df3['t'], y=df3['ln'], s=0.3, label='2: 0.012M $\ NaOH$')
plt.scatter(x=df4['t'], y=df4['ln'], s=0.3, label='3: 0.05M $\ NaOH$')
plt.scatter(x=df5['t'], y=df5['ln'], s=1, label='4: 0.075M $\ NaOH$')

plt.xlabel('t, c')
plt.ylabel('$ln\left(\\frac{D}{D_0}\\right)$')
plt.legend()
plt.savefig('1ln.png')
plt.show()
```

```python
plt.figure(figsize=(12,7))

df1_ = df1[df1['t'] < 65]
k, q = np.polyfit(list(df1_['t']), list(df1_['ln']), deg=1)
sns.lineplot(data=df1_, x='t', y='ln', label='5: 0.1M $\ NaOH\ \ \ :\ \ y = %.4f\\cdot x + %.4f\ $' %(k, q))

df2_ = df2
k, q = np.polyfit(list(df2_['t']), list(df2_['ln']), deg=1)
sns.lineplot(data=df2_, x='t', y='ln', label='1: 0.003M $\ NaOH\ \ \ :\ \ y = %.4f\\cdot x %.4f\ $' %(k, q))

df3_ = df3
k, q = np.polyfit(list(df3_['t']), list(df3_['ln']), deg=1)
sns.lineplot(data=df3_, x='t', y='ln', label='2: 0.012M $\ NaOH\ \ \ :\ \ y = %.4f\\cdot x %.4f\ $' %(k, q))

df4_ = df4[df4['t'] < 100]
k, q = np.polyfit(list(df4_['t']), list(df4_['ln']), deg=1)
sns.lineplot(data=df4_, x='t', y='ln', label='3: 0.05M $\ NaOH\ \ \ :\ \ y = %.4f\\cdot x %.4f\ $' %(k, q))

df5_ = df5[df5['t'] < 80]
k, q = np.polyfit(list(df5_['t']), list(df5_['ln']), deg=1)
sns.lineplot(data=df5_, x='t', y='ln', label='4: 0.075M $\ NaOH\ \ \ :\ \ y = %.4f\\cdot x %.4f\ $' %(k, q))

plt.xlabel('t, c')
plt.ylabel('$ln\left(\\frac{D}{D_0}\\right)$')
plt.legend()
plt.savefig('1linear_ln.png')
plt.show()
```

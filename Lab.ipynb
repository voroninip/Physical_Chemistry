```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_excel(io='коту2.xlsx', sheet_name='Лист1')
df
```

```python
plt.figure(figsize=(12,8))

k, q = np.polyfit(list(df['dP1']), list(df['dT1']), deg=1)
sns.lineplot(data=df, x='dP1', y='dT1', color='red')
sns.scatterplot(data=df, x='dP1', y='dT1', color='red')
x=df['dP1']
y=k*x + q
plt.plot(x, y, label='$19.6C^{\circ}\ :\ \ y = %.4f\\cdot x %.4f\ $' %(k, q), color='pink')

k, q = np.polyfit(list(df['dP2']), list(df['dT2']), deg=1)
sns.lineplot(data=df, x='dP2', y='dT2', color='green')
sns.scatterplot(data=df, x='dP2', y='dT2', color='green')
x=df['dP2']
y=k*x + q
plt.plot(x, y, label='$30.0C^{\circ}\ :\ \ y = %.4f\\cdot x %.4f\ $' %(k, q), color='lime')

k, q = np.polyfit(list(df['dP3']), list(df['dT3']), deg=1)
sns.lineplot(data=df, x='dP3', y='dT3', color='blue')
sns.scatterplot(data=df, x='dP3', y='dT3', color='blue')
x=df['dP3']
y=k*x + q
plt.plot(x, y, label='$40.0C^{\circ}\ :\ \ y = %.4f\\cdot x + %.4f\ $' %(k, q), color='cyan')

plt.xlabel('$\\Delta P, атм.$')
plt.ylabel('$\\Delta T,K$')
plt.legend()
plt.savefig('approx.png')
plt.show()
```
